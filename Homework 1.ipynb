```python
#Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân. 
a = float(input("chieu rong a: "))

b = float(input("chieu rong b: "))

pi = 3.14
r = a / 2 

S_rect = a * b
S_circle = pi * r * r

S = S_rect - S_circle 
print(round(S, 2))
```

```python
#Bài 3: Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem trước cấu trúc lựa chọn: if….else, ma) 
c = input("Nhập một kí tự: ")

if c.isupper():
    print(c.lower())
elif c.islower():
    print(c.upper())
else:
    print("Kí tự không phải chữ cái")
```

```python
#Bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương ứng của A. Xét TH khi c = ‘A’ là đặc biệt
c = input("Nhập chữ cái hoa: ")

c_thuong = chr(ord(c) + 32)

if c_thuong == 'a':
    print('z')
else:
    print(chr(ord(c_thuong) - 1))
```

```python
#Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.
import math
a = float(input("Nhap a: "))
b = float(input("Nhap b: "))
c = float(input("Nhap c: "))
if a + b > c and a + c > b and b + c > a:
    print("a b c la canh cua 1 tam giac")
    p = (a + b + c) / 2
    s = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"Dien tich tam giac: {s:.1f}")      
else:
    print("khong phai canh cua 1 tam giac")
```

```python
#6. Kiểm tra ký tự nhập vào là chữ hay số: Viết chương trình nhập một ký tự, kiểmtra xem ký tự đó là chữ cái hay số.
ch = input("Nhập một ký tự: ")

if ch.isalpha():
    print("Đây là chữ cái")
elif ch.isdigit():
    print("Đây là chữ số")
else:
    print("Ký tự khác")
```

```python
#7. Bài tập 5: Tính điểm học lực Nhập điểm (0-10), phân loại học lực theo quy tắc: >=8: Giỏi; >=6.5 và <8: Khá; >=5 và <6.5: Trung bình; <5: Yếu
n = float(input())
if n < 0:
    print(False)
elif n < 5:
    print("Yếu")
elif 5 <= n < 6.5:
    print("Trung bình")
elif 6.5 <= n < 8:
    print("Khá")
else:
    print("Giỏi")
```

```python
#8. Kiểm tra năm nhuận: Viết chương trình nhập năm, kiểm tra xem năm đó có phải năm nhuận hay không. (Năm nhuận nếu chia hết cho 400 hoặc chia hết cho 4nhưng không chia hết cho 100) 

n = int(input())
if (n % 4 == 0 and n % 100 != 0) or n % 400 == 0:
    print(n, "là năm nhuận")
else:
    print(n, "là năm bình thường")
```

```python
#9. Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case 

n = int(input())
match n:
    case 0:
        print("so khong")
    case 1:
        print("so mot")
    case 2:
        print("so hai")
    case 3:
        print("so ba")
    case _:
        print("ERor")
```
