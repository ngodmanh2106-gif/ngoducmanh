W5A1 Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) ->
int:

```python
def max_of_two(a, b):
    return max(a, b)
a, b = map(int, input().split())
print(max_of_two(a, b))
```

W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không?


```python
def snt(a):
    if a < 2:
        return False
    for i in range(2, int(a**0.5) + 1):
        if a % i == 0:
            return False
        return True
a = int(input())
if snt(a):
    print(a, "là số nguyên tố")
else:
    print(a, "không là số nguyên tố")

    
        

    
 
```

W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 


```python
def so_hoan_hao(n):
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    if tong == n:
        return n, "là số hoàn hảo"
    else:
        return n, "không là số hoàn hảo"
n = int(input())
print(so_hoan_hao(n))



```

W5A5
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử
đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 

```python
def tim_k(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1
k = int(input())
lst = list(map(int, input().split()))
print(tim_k(lst, k))
```

W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 


```python
def giai_thua(n):
    tich = 1
    for i in range(1, n + 1):
        tich *= i
    return tich
n = int(input())
print(giai_thua(n))
```

Viết hàm giả lập một máy tính bỏ túi.
Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat
\)
Biết các toán tử đơn giản thực hiện phép toán lần lượt là:
Phép trừ \( - \)
Phép cộng \( + \)
Phép chia \( / \)
Phép nhân \( * \)
Kết quả làm tròn đến chữ số thập phân thứ hai.
Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)).
Đầu ra: Kết quả phép toán

```python
def casio(num1, num2, operat):
   if operat == "+":
      return round(num1 + num2, 2)
   elif operat == "-":
      return round(num1 - num2, 2)
   elif operat == "*":
      return round(num1 * num2, 2)
   elif operat == "/":
      if num1 == 0:
         return "không chia được"
      return round(num1 / num2, 2)
num1, num2 = map(int, input().split())
operat = input()
print(casio(num1, num2, operat))
```

W5A8 Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. 

```python
def hamming(a, b):
    x = a ^ b 
    tong = 0 
    while x > 0:
        if x % 2 == 1:
            tong += 1
        x = x // 2
    return tong
a, b = map(int, input().split())
print(hamming(a, b))
```

W5A9 Viết hàm trả về tổng các chữ số của số nguyên dương n 


```python
def sum_num(n):
    tong = 0
    while n > 0:
        tong += n % 10
        n //= 10 
    return tong
n = int(input())
print(sum_num(n))
```

W5A11 Viết hàm is_product_divisible_by_10(a, b) trả về True nếu tích a*b chia hết cho 10, ngược lại False.

```python
def is_product_divisible_by_10(a, b):
    if a * b % 10 == 0:
        return True
    else:
        return False
a, b = map(int, input().split())
print(is_product_divisible_by_10(a, b))
```

W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông

```python

def is_right_triangle(a, b, c):
    if a*a == (b*b + c*c) or b*b == (a*a + c*c) or c*c == (b*b + a*a):
        return True
    else:
        return False
a, b, c = map(int, input().split())
print(is_right_triangle(a, b, c))
```
