```python
#Bài 3:
import math

a, b, c = map(float, input().split())
if a + b <= c and a + c <= b and b + c <= a:
    print("a b c khong la canh cua 1 tam giac")
else:
    if a == b == c:
        print("đây là tam giác đều")
    elif a*a + b*b == c*c or a*a + c*c == b*b or b*b + c*c == a*a:
        print("đây là tam giác vuông")
    elif a == b or a == c or c == b:
        print("đây là tam giác cân")
    else:
        print("đây là tam giác thường")
# chu vi 
P = a + b + c 
print("Chu vi là:", P)
p = P/2
S = math.sqrt(p * (p - a) * (p - b) * (p - c))
print("Diện tích là: ", f"{S:.1f}")
```

```python
#Bài 5:
import math
def hpt_bachai(a, b, c):
    d = b*b - 4*a*c
    if a == 0 and b == 0:
        return "vo nghiem"
    elif a == 0:
        return "phuong trinh co 1 nghiem duy nhat", -c / b
    else:
        if d < 0:
            return "vo nghiem"
        elif d == 0:
            return "phuong trinh co nghiem kep r1 = r2", -b / (2*a)
        else:
            r1 = (-b + math.sqrt(d)) / 2*a
            r2 = (-b - math.sqrt(d)) / 2*a 
            return "phuong trinh co 2 nghiem", r1, r2
print(hpt_bachai(0, 6, 4))
```
