```python
#W3A11
import math

a, b, c = map(float, input().split())
if a + b <= c and a + c <= b and b + c <= a:
    print("a b c khong la canh cua 1 tam giac")
else:
    if a == b == c:
        print("đây là tam giác đều")
    elif a*a + b*b == c*c or a*a + c*c == b*b or b*b + c*c == a*a:
        print("đây là tam giác vuông")
    elif a == b or a == c or c == b:
        print("đây là tam giác cân")
    else:
        print("đây là tam giác thường")
# chu vi 
P = a + b + c 
print("Chu vi là:", P)
p = P/2
S = math.sqrt(p * (p - a) * (p - b) * (p - c))
print("Diện tích là: ", f"{S:.1f}")
```

```python
#W3A17
a = int(input("Nhập a: "))
b = int(input("Nhập b: "))
c = int(input("Nhập c: "))
d = int(input("Nhập d: "))


if a == 0:
    if b == 0 and c == 0 and d == 0:
        print("YES")
    else:
        print("NO")
else:
    if b % a != 0:
        print("NO")
    else:
        q = b // a
        if b * q == c and c * q == d:
            print("YES")
        else:
            print("NO")
```
