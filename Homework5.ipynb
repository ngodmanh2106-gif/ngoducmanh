Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def armstrong(x):
    n = abs(x)
    tong = 0
    k = len(str(n))
    t = n
    while t > 0:
        s = t % 10
        tong += s ** k
        t //= 10
    return tong == n

x = int(input())

print(armstrong(x)) 
```

Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x


```python
def chinh_phuong(x):
    if x < 0:
        return False
    i = 0
    while i * i <= x:
        if i * i == x:
            return True
        i += 1
    return False
x = int(input())
print(chinh_phuong(x))
```

W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def palindrome(x):
    if x < 0:
        return False
    org = x
    rev = 0
    while x > 0:
        d = x % 10
        rev = rev * 10 + d
        x //= 10
    return rev == org
x = int(input())
print(palindrome(x))
```

W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.


```python
def wonder(x):
    seen = set()
    
    while x != 1:
        if x in seen:
            return False
        
        seen.add(x)
        
        tong = 0
        while x > 0:
            d = x % 10
            tong += d * d
            x //= 10
        
        x = tong
    
    return True
x = int(input())
print(wonder(x))
```

W5A21 Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại

```python
def is_one_product_of_others(a, b, c):
    if a == b * c or b == a * c or c == a * b:
        return True
    else:
        return False
a, b, c = map(int, input().split())
print(is_one_product_of_others(a, b, c))
```

W5A22 Viết hàm sum_halves(s) nhận chuỗi 6 chữ số và trả về tổng của hai số 3 chữ số đầu và 3 chữ số cuối.

```python
def sum_halves(s):
    if len(s) != 6:
        return None
    else:

        first = int(s[:3])
        last = int(s[3:])
    return first + last
s = input()
print(sum_halves(s))
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.


```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    area = x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)
    return area == 0
x1, y1, x2, y2, x3, y3 = map(int, input().split())
print(are_collinear(x1, y1, x2, y2, x3, y3))
```

W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)


```python
def khac0(s):
    tong = 0
    for i in (s):
        i = abs(i)
        if i == 0:
            continue
        while i > 0:
            d = i % 10
            if d != 0:
                tong += 1
            i //= 10
    return tong
s = list(map(int, input().split()))
print(khac0(s))
```

W5A29 Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min].


```python
def max_min(a, b, c):
    return [max(a, b, c), min(a, b, c)]
    
a, b, c = map(int, input().split())
print(max_min(a, b, c))
```

W5A31 Viết hàm tính tổng bình phương các số từ 1 đến n.

```python
def sum_double(n):
    tong = 0
    for i in range(1, n + 1):
        tong += i*i
    return tong
n = int(input())
print(sum_double(n))
```
