```python
#3.4 
# Nhập chiều cao h từ bàn phím
h = int(input("Nhập chiều cao h = "))

# Vòng lặp chạy qua từng dòng, từ dòng 1 đến dòng h
for i in range(1, h + 1):
    
    # 1. Vòng lặp in khoảng trắng ở đầu mỗi dòng (để tạo độ nghiêng)
    for j in range(h - i):
        print(" ", end="")
        
    # 2. Xử lý in dấu * và khoảng trắng bên trong tam giác
    if i == 1:
        # Dòng đầu tiên (đỉnh): chỉ in 1 dấu *
        print("*")
        
    elif i == h:
        # Dòng cuối cùng (đáy): in toàn bộ dấu * (số lượng = 2*h - 1)
        for j in range(2 * h - 1):
            print("*", end="")
        print() # Xuống dòng sau khi in xong đáy
        
    else:
        # Các dòng ở giữa: viền trái -> khoảng trống rỗng -> viền phải
        print("*", end="")
        
        # Vòng lặp in khoảng trắng rỗng bên trong
        # Số khoảng trắng luôn tuân theo công thức (2*i - 3)
        for j in range(2 * i - 3):
            print(" ", end="")
            
        print("*")
```

```python
#3.8
a, b = map(int, input().split())

if a <= 0 or b <= 0:
    print("error!")
else:
    limit = min(a , b)
    for i in range(1, limit + 1):
        if a % i == 0 and b % i == 0:
            print(i, end=" ")
```

```python
#3.9
a = int(input("Nhập số nguyên dương a: "))

if a < 0:
    print("Không phải số chính phương")
else:
    i = 0
    while i * i <= a:
        if i * i == a:
            print(a, "là số chính phương")
            break
        i += 1
    else:
        print(a, "không phải số chính phương")
```

```python
#3.10
a = int(input("Nhập số nguyên dương a: "))

tong = 0
for i in range(1, a):
    if a % i == 0:
        tong += i

if tong == a:
    print(a, "là số hoàn hảo")
else:
    print(a, "không phải số hoàn hảo")
```
